```python
import pandas as pd

# Criando um DataFrame maior
data = {
    'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Fay', 'George', 'Hannah', 'Ian', 'Jane'],
    'Idade': [25, 30, 35, 40, 45, 28, 33, 38, 43, 27],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba', 'Porto Alegre', 'São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba', 'Porto Alegre'],
    'Salário': [5000, 6000, 7000, 8000, 9000, 5500, 6500, 7500, 8500, 9500]
}

df = pd.DataFrame(data)
display(df)
```

### head(): Mostra as primeiras 5 linhas do DataFrame (por padrão).

```python
display(df.head())
```

### tail(): Mostra as últimas 5 linhas do DataFrame (por padrão).

```python
display(df.tail())
```

### info(): Fornece um resumo das informações do DataFrame, como tipos de dados e valores nulos.

```python
display(df.info())
```

### describe(): Gera estatísticas descritivas para colunas numéricas.

```python
display(df.describe())
```


### value_counts(): Conta quantas vezes aparecem as ocorrências de valores únicos em uma coluna.

```python
display(df['Cidade'].value_counts())
```

### loc[]: Seleciona linhas e colunas com base em rótulos.

```python
display(df.loc[2])         # Linha com índice 2
display(df.loc[:, 'Nome']) # Coluna 'Nome'
```

### iloc[]: Seleciona linhas e colunas com base em índices inteiros.

```python
display(df.iloc[2])        # Linha com índice 2
display(df.iloc[:, 0])    # Primeira coluna
```

### sort_values(): Ordena o DataFrame por uma ou mais colunas.

```python
display(df.sort_values(by='Idade'))
```

### drop(): Remove linhas ou colunas do DataFrame.

```python
display(df.drop(columns='Salário'))
```

### apply(): Aplica uma função a cada elemento de uma coluna ou linha.

```python
display(df['Salário'].apply(lambda x: x * 1.1))  # Aumenta o salário em 10%
```

### rename(): Renomeia colunas ou índices.

```python
df_renomeado = df.rename(columns={'Nome': 'Nome Completo', 'Idade': 'Idade Anos'})
display(df_renomeado)
```

### fillna(): Preenche valores nulos com um valor específico.

```python
df_completo = df.fillna({'Salário': df['Salário'].mean()})
display(df_completo)
```

### dropna(): Remove linhas ou colunas com valores nulos.

```python
df_sem_nulos = df.dropna()
display(df_sem_nulos)
```

### groupby(): Agrupa dados com base em uma ou mais colunas e aplica uma função agregadora.)


```python
df_agrupado = df.groupby('Cidade').min()
display(df_agrupado)
```

### merge(): Mescla dois DataFrames com base em uma coluna comum

```python
data_adicional = {
    'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Departamento': ['Financeiro', 'RH', 'TI', 'Marketing', 'Vendas']
}
df_adicional = pd.DataFrame(data_adicional)
df_completo = pd.merge(df, df_adicional, on='Nome')
display(df_completo)
```

### astype(): Converte o tipo de dados de uma coluna.

```python
df['Idade'] = df['Idade'].astype(float)
display(df)
```

### duplicated(): Identifica linhas duplicadas no DataFrame.

```python
display(df[df.duplicated()])
```
