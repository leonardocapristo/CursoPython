### Importando o pandas

```python
import pandas as pd
```

### Criando um dataframe a partir de um dicionário

```python
# dataframe = pd.DataFrame()
Teste = {'data': ['15/02/2021', '16/02/2021'],
         'valor': [500, 300],
         'produto': ['feijao', 'arroz'],
         'qtde': [50, 70],
        }

Teste_df = pd.DataFrame(Teste)
```

### Visualização dos Dados

```python
print(Teste_df)
display(Teste_df)
```

### Importando arquivos e bases de dados
